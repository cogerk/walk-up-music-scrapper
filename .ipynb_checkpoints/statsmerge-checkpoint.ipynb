```python
import pybaseball
from itertools import zip_longest
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
sns.set(color_codes=True)
```

*DON'T RUN* Retrieve most recent stats

```python
df_batting = pybaseball.batting_stats(2018)
data_dir = 'data'
fname_batting = '2018_stats_batting.csv'
df_batting.to_csv(data_dir + '/' + fname_batting, index=False)

df_pitching = pybaseball.pitching_stats(2018)
fname_pitching = '2018_stats_pitching.csv'
df_pitching.to_csv(data_dir + '/' + fname_pitching, index=False)
```

Load in walk up music database & combine stats & walk up music + do data cleaning

```python
df_music = pd.read_csv('data/WalkUpMusic_Discogs_Complete.csv')
df_ext = pd.read_csv('data/player_ages_ext.csv')

df_music['Player_Name'], df_music['Situation'] = list(zip_longest(*[[y.replace(')', '') 
                                                         for y in x.Player_Name.split(' (')] 
                                                                    for idx, x in 
                                                                    df_music.iterrows()]))
df_batting = pd.read_csv('data/2018_stats_batting.csv')
df_pitching = pd.read_csv('data/2018_stats_pitching.csv')
df = df_music.merge(df_batting, 
                    how='left', 
                    left_on='Player_Name', 
                    right_on='Name').merge(df_pitching,
                                          how='left',
                                          left_on='Player_Name',
                                          right_on='Name').merge(df_ext, 
                                                                 how='left', 
                                                                 on='Player_Name')

df =  df.assign(Age = [x.Age_x if not np.isnan(x.Age_x) else 
                       x.Age_y if not np.isnan(x.Age_y) else x.Age_ext
                       for idx, x in df.iterrows()]).drop(
    columns =['Age_x','Age_y', 'Age_ext','Season_x', 'Season_y', 'Team_x', 'Team_y', 'Name_x', 'Name_y'])

df = df.assign(Age_On_Release = df['Age']  - 2018 + df['Release_Year'])
```

```python
df.to_csv(data_dir + '/' + 'all_data.csv', index=False)
```
